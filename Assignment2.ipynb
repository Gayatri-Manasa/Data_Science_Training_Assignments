```python
# 1. Basic String Manipulation
# Assign the string "Hello, World!" to a variable and print it
greeting = "Hello, World!"
print(greeting)
# Create a string "Python Programming" and extract the first five characters using slicing
course = "Python Programming"
print(course[:5])
```

```python
# 2. String Methods
# Given a string "python is fun", convert it to uppercase
text = "python is fun"
print(text.upper())
# Replace "fun" with "awesome" in the string "Python is fun"
new_text = "Python is fun".replace("fun", "awesome")
print(new_text)
```

```python
# 3. String Formatting
# Use f-string formatting for name and age
name = "Gayatri"
age = 25
print(f"My name is {name}, and I am {age} years old.")
# Given a price of 49.99, format it to display only two decimal places
price = 49.99
print(f"The price is ${price:.2f}")
```

```python
# 4. String Functions
# Count the occurrences of the letter 'o' in "Hello, how are you?"
sentence = "Hello, how are you?"
print("Number of 'o's:", sentence.count('o'))
# Find the position of "world" in "Hello, world! Python is amazing."
phrase = "Hello, world! Python is amazing."
position = phrase.find("world")
print("Position of 'world':", position)
```

```python
# 5. Reverse and Check
# Reverse the string "Python" using slicing
word = "Python"
print("Reversed:", word[::-1])
# Check if "madam" is a palindrome
word2 = "madam"
if word2 == word2[::-1]:
print(f"'{word2}' is a palindrome.")
else:
print(f"'{word2}' is not a palindrome.")
```
