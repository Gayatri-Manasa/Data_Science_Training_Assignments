```python
##Basic Array Operaions
#1.Create a 3x3 array with random integers, calculate sum, mean, and standard deviation
import numpy as np
arr1 = np.random.randint(0, 11, (3, 3))
print("Array:\n", arr1)
print("Sum:", np.sum(arr1))
print("Mean:", np.mean(arr1))
print("Standard Deviation:", np.std(arr1))
```

```python
#3.Generate two 2x3 arrays and perform element-wise operations.arr3 = np.random.randint(1, 11, (2, 3))
arr4 = np.random.randint(1, 11, (2, 3))
print("Array 1:\n", arr3)
print("Array 2:\n", arr4)
print("Addition:\n", arr3 + arr4)
print("Subtraction:\n", arr3 - arr4)
print("Multiplication:\n", arr3 * arr4)
print("Division:\n", arr3 / arr4)
```

```python
#8.Stack two 3x3 matrices horizontally and vertically.
arr8_1 = np.random.randint(1, 11, (3, 3))
arr8_2 = np.random.randint(1, 11, (3, 3))
print("Horizontal Stack:\n", np.hstack((arr8_1, arr8_2)))
print("Vertical Stack:\n", np.vstack((arr8_1, arr8_2)))
```

```python
#9.Concatenate two arrays of different sizes along a new axis.
arr9_1 = np.random.randint(1, 11, 3)
arr9_2 = np.random.randint(1, 11, 5)
# Concatenate along axis=0 to stack vertically,
# since the arrays have different lengths but can be stacked on top of each other
concatenated_arr = np.concatenate((arr9_1[:, np.newaxis], arr9_2[:, np.newaxis]), axis=0)
print(concatenated_arr)
```

```python
##Linear Alzebra
#26.Create a 3x3 matrix and find its determinant.
arr26 = np.random.randint(1, 11, (3, 3))
print("Matrix:\n", arr26)
print("Determinant:", np.linalg.det(arr26))
```

```python
#27.Compute the inverse of a 2x2 matrix and verify.
arr27 = np.random.randint(1, 11, (2, 2))
inverse_arr27 = np.linalg.inv(arr27)
identity_check = np.dot(arr27, inverse_arr27)
print("Original Matrix:\n", arr27)
print("Inverse:\n", inverse_arr27)
print("Identity Check:\n", np.round(identity_check))
```

```python
#28.Calculate eigenvalues and eigenvectors of a 2x2 matrix.
arr28 = np.random.randint(1, 11, (2, 2))
eigenvalues, eigenvectors = np.linalg.eig(arr28)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
#29.Solve the system of linear equations 2x + 3y = 5 and 4x + 6y = 10.
A= np.array([[2, 3], [4, 6]])
B = np.array([5, 10])
try:
    # The following two lines were not indented properly
    solution = np.linalg.solve(A, B)
    print("Solution:", solution)
except np.linalg.LinAlgError:
    # This line was also not indented properly
    print("No unique solution, system is singular.")
```

```python
#30.Perform Singular Value Decomposition (SVD) and reconstruct matrix.
arr30 = np.random.randint(1, 11, (3, 3))
U, S, Vt = np.linalg.svd(arr30)
S_matrix = np.diag(S)
reconstructed_matrix = np.dot(U, np.dot(S_matrix, Vt))
print("Original Matrix:\n", arr30)
print("Reconstructed Matrix:\n", np.round(reconstructed_matrix))
```
