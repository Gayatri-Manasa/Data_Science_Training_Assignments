```python
#1.Tuple Creation
my_tuple = (10, "apple", 3.14, "banana", 42)
print("Tuple:", my_tuple)
#2.Accessing Elements
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
print("Second last element (negative indexing):", my_tuple[-2])
#3.Tuple Slicing
print("First three elements:", my_tuple[:3])
print("Last two elements:", my_tuple[-2:])
#4.Tuple Operations
another_tuple = ("grape", 99)
concatenated = my_tuple + another_tuple
print("Concatenated tuple:", concatenated)
repeated = my_tuple * 2
print("Repeated tuple:", repeated)
#5.Tuple Methods
index_value = my_tuple.index("banana")
print("Index of 'banana':", index_value)
count_value = my_tuple.count(10)
print("Count of 10:", count_value)
#6.Tuple Immutability
try:
  my_tuple[1] = "orange"
except TypeError as e:
  print("Error:", e)

print("Explanation: Tuples are immutable, meaning their elements cannot be changed once created.")
#7.Tuple Packing and Unpacking
person = ("Alice", 28, "Developer")
name, age, job = person
print("Name:", name)
print("Age:", age)
print("Job:", job
)
#8.Tuple Iteration
print("Iterating through my_tuple:")
for item in my_tuple:
  print(item)
#9.Tuple Usage - Function returning multiple values as a tuple
def calculate_values(a, b):
  return (a + b, a - b, a * b, a / b)
result = calculate_values(20, 5)
print("Returned tuple from function:", result)
```
